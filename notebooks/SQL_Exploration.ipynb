### User history

**At User Level**   
No. Of Prior PhoneRequests   
No. Of Prior LandingPage Views   
No. Of Prior Searches   
Last Searchdate   
First Searchdate   
No. Of Days between first and last search date

```python
# USER number of landing page views
query = """SELECT user_id, count(ad_id) num_land_page_views 
           FROM train_visits_stream 
           GROUP BY user_id 
           ORDER BY num_land_page_views DESC;"""

landing_page_views = pd.read_sql(query, connection)
print(landing_page_views.shape)
print('landing_page_views')
landing_page_views.head()
```

```python
# USER number of phone requests
query = """SELECT user_id, count(ad_id) num_phone_req 
           FROM train_phone_requests_stream 
           GROUP BY user_id 
           ORDER BY num_phone_req DESC;"""

num_phone_req = pd.read_sql(query, connection)
print(num_phone_req.shape)
print('num_phone_req')
num_phone_req.head()
```

```python
# USER number of searches
query = """SELECT 
            user_id,
            count(search_id) num_searches,
            min(search_date) first_search_date, 
            max(search_date) last_search_date,
            (max(search_date)::date - min(search_date)::date) num_days
           FROM train_search_info 
           GROUP BY user_id 
           ORDER BY num_searches DESC;"""

num_searches = pd.read_sql(query, connection)
print(num_searches.shape) #514568
print('num_searches')
num_searches.head()
```

#### Merge user info together

```python
user_info = (num_searches
             .merge(landing_page_views, how='left', on='user_id')
             .merge(num_phone_req,how='left', on='user_id')
             .fillna(0))

print(user_info.shape)
user_info.head()
```

### Search Info and details

```python
query = """SELECT 
                train_search_stream.search_id,
                train_search_stream.is_click,
                train_search_stream.ad_id,
                train_search_stream.position,
                train_search_stream.hist_ctr,
                category.subcategory category,
                extract(dow from search_date) day_of_week,
                extract(hour from search_date) hour_of_day,
                train_ads_info.price,
                train_ads_info.title,
                location.region_id region,
                train_search_info.user_id
            FROM train_search_stream 
            LEFT JOIN train_search_info
                ON train_search_stream.search_id = train_search_info.search_id
            LEFT JOIN train_ads_info
                ON train_search_stream.ad_id = train_ads_info.ad_id
            LEFT JOIN category
                ON train_ads_info.category_id = category.category_id
            LEFT JOIN location
                ON train_search_info.search_location_id = location.location_id
            GROUP BY 
                train_search_stream.search_id, 
                train_search_stream.ad_id, 
                train_search_stream.position, 
                train_search_stream.object_type, 
                train_search_stream.hist_ctr, 
                train_search_stream.is_click,
                train_search_info.user_id,
                train_ads_info.category_id,
                train_ads_info.price,
                category.subcategory,
                train_search_info.search_date,
                location.region_id,
                train_ads_info.title
            """

search_info = pd.read_sql(query, connection)
print(search_info.shape) 
print('search_info')
search_info.head()
```

## Merge for the final data frame for modeling

```python
df = search_info.merge(user_info, how='left', on='user_id')
print(df.shape)
df.head()
```

```python
df['num_days'] = [x+1 for x in df['num_days']]
df = df.dropna()
print(df.shape)
df.head()
```
